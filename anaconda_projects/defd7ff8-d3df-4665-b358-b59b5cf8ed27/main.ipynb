```python
def convert_to_binary(myNum):
    """
    Convert from base-n to binary, assume default base10
    I actually already wrote this for this class, because I misunderstood
    The first test and thought we had to convert to binary in Python :-)
    Now I just need to find it...
    Here's how you search for "that one python file I used the divmod() command in",
    on a *nix system...
    (base) ~preece$ find . -iname '*.py' -exec grep -Hi divmod {} \;
    With enough cli skills, you can work around your own disorginization :-D
    """
    mylist = []
    myint, remainder = divmod(myNum,2)
    while not ((myint == 0) and (remainder == 0)):
        mylist.append(remainder)
        myint, remainder = divmod(myint, 2)
    mylist.reverse() # reverse the list
    # I forget this, every time: https://stackoverflow.com/questions/19299563/list-reverse-is-not-working
    # "reverse modifies the list in place and returns None."
    mylist = [str(x) for x in mylist] # anonymous function. We're casting ints to strs, *within* the list itself
    # it now looks like this: ['1', '0', '1', '0', '1', '0'] a list of strings, not ints
    myString = "".join(mylist) # now, squash it down to a string: '101010'
    return int(myString) # finally, cast it *back* to int. The string manipulation was just to strip the list formatting
    # Probably better to represent it as a binary number like 0b101010, but to keep it simple we're just using int()
```
